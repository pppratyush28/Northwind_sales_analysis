```python
# Northwind Data Analysis â€” Visual Insights

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the visual style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# Monthly Revenue Trend

monthly_df = pd.read_csv("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/CSV Exports/Monthly_Revenue_Trend.csv")
monthly_df['month'] = pd.to_datetime(monthly_df['month'])

plt.figure()
sns.lineplot(data=monthly_df, x='month', y='revenue', marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/visuals/monthly_revenue.png")
plt.show()
```

```python
# Top Customers by Revenue 

top_customers = pd.read_csv("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/CSV Exports/Top_Spending_Customers.csv")

plt.figure()
sns.barplot(
    data=top_customers,
    y='name_of_company',
    x='total_spend',
    hue='name_of_company',      # add hue to match each bar with a color
    dodge=False,
    palette='rocket',
    legend=False                # suppress legend since y-axis labels are already descriptive
)
plt.title("Top Customers by Revenue")
plt.xlabel("Revenue ($)")
plt.ylabel("Customer")
plt.tight_layout()
plt.savefig("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/visuals/top_customers_bar.png")
plt.show()
```

```python
# Top Products by Revenue

top_products = pd.read_csv("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/CSV Exports/Top_Products_by_Revenue.csv")
top10 = top_products.head(10)

plt.figure(figsize=(10, 10))
plt.pie(top10['total_revenue'], labels=top10['product_name'], autopct='%1.1f%%', startangle=135)
plt.title("Top 10 Products by Revenue")
plt.tight_layout()
plt.savefig("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/visuals/product_revenue_pie.png")
plt.show()
```

```python
# Customer Segments by Lifetime Value

segments_df = pd.read_csv("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/CSV Exports/Customer_Segmentation.csv")
segment_counts = segments_df['customer_segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'Customer Count']

plt.figure()
sns.barplot(data=segment_counts, x='Segment', y='Customer Count', hue = 'Customer Count', palette='bright')
plt.title("Customer Segments by Lifetime Value")
plt.xlabel("Segment")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.savefig("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/visuals/customer_segments_chart.png")
plt.show()
```

```python
# Repeat Customers Count

repeat_df = pd.read_csv("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/CSV Exports/No_of_repeat_customers.csv")
labels = ['Repeat Customers', 'One-time Customers']
values = [repeat_df['repeat_customers'].iloc[0], repeat_df['single_time_visitors'].iloc[0]]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#4caf50', '#ff9800'])
plt.title("Repeat vs One-Time Customers")
plt.tight_layout()
plt.savefig("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/visuals/repeat_customers_pie.png")
plt.show()
```

```python
# Average Order Value by Customer

aov_df = pd.read_csv("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/CSV Exports/avg_order_value.csv")

plt.figure()
sns.histplot(data=aov_df, x='avg_order_value', bins=20, kde=True, color='dodgerblue')
plt.title("Average Order Value Distribution")
plt.xlabel("Average Order Value ($)")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.savefig("/Users/pratyushpranjal/Documents/GitHub/Northwind_sales_analysis/visuals/aov_distribution.png")
plt.show()
```
